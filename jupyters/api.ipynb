```python
import pandas as pd
import requests
```

```python
url = "https://datos.madrid.es/egob/catalogo/300107-0-agenda-actividades-eventos.json"
```

```python
response = requests.get(url)
data = response.json()
```

```python
eventos = data.get('@graph', [])

df = pd.DataFrame(eventos)
df.head()
```

```python
print(df['address'].head())
```

```python
df = df[['title', 'link', 'address', 'event-location', 'dtstart', 'dtend', 'organization']]
df.head()
```

```python
df['dtstart'] = pd.to_datetime(df['dtstart'], errors='coerce')
df['dtend'] = pd.to_datetime(df['dtend'], errors='coerce')
fecha_inicio = "2025-03-01"
fecha_fin = "2025-03-02"
df = df[(df['dtstart'] <= fecha_inicio) & (df['dtend'] >= fecha_fin)]
df.head()
```

```python
def extraer_direccion(info):
    if isinstance(info, dict):
        # Revisar si hay un campo 'area' con la información relevante
        area = info.get('area', {})
        direccion = area.get('street-address', None)  # Dirección
        codigo_postal = area.get('postal-code', None)  # Código postal
        return direccion, codigo_postal
    return None, None

```

```python
df[['direccion', 'codigo_postal']] = df['address'].apply(lambda x: pd.Series(extraer_direccion(x)))
```

```python
def extraer_organizacion(info):
    if isinstance(info, dict):
        # Extraer el nombre de la organización
        return info.get('organization-name', None)
    return None
df['organizacion'] = df['organization'].apply(lambda x: extraer_organizacion(x))
```

```python
df = df.rename(columns={
    'title': 'nombre_evento',
    'link': 'url_evento',
    'dtstart': 'fecha_inicio',
    'dtend': 'fecha_fin',
    'event-location': 'localizacion',
    'organization': 'organizacion'
})
```

```python
df['codigo_postal'] = pd.to_numeric(df['codigo_postal'], errors='coerce')
```

```python
df.head()
```

```python
df = df.drop(columns=['address'])
df['codigo_postal'] = df['codigo_postal'].fillna(0).astype(int)
df.head()
```

```python
df = df[['nombre_evento', 'url_evento', 'codigo_postal', 'direccion', 'fecha_inicio', 'fecha_fin', 'localizacion']]
```

```python
df.head()
```

```python
df['horario'] = df['fecha_inicio'].dt.strftime('%H:%M') + ' - ' + df['fecha_fin'].dt.strftime('%H:%M')
df.head()
```

```python
df['fecha_inicio'] = df['fecha_inicio'].dt.date
df['fecha_fin'] = df['fecha_fin'].dt.date
df.head()
```

```python
df.info()
```

```python
df.to_parquet('../data/eventos_nuevos.parquet', index=False)
```
