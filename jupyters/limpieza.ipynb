```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_parquet('../data/reservas_hoteles.parquet')
```

```python
df_raw.info()
```

```python
df = df_raw
```

```python
df_raw['nombre_hotel'].unique()
```

```python
df_raw['inicio_estancia'] = '2025-03-01'
df_raw['final_estancia'] = '2025-03-02'
df_raw.head()
```

```python
df_raw['ciudad'] = 'Madrid'
```

```python
valores_unicos_id_competencia = df_raw.loc[df_raw['nombre_hotel'] == '', 'id_hotel'].unique()
valores_unicos_id_competencia
```

```python
df = df_raw
df.info()
```

```python
media_estrellas = df[df['competencia'] == False].groupby('nombre_hotel')['estrellas'].transform('mean')

df.loc[df['competencia'] == False, 'estrellas'] = media_estrellas

media_precio = df[df['competencia'] == False].groupby('nombre_hotel')['precio_noche'].transform('mean')

df.loc[df['competencia'] == False, 'precio_noche'] = df.loc[df['competencia'] == False, 'precio_noche'].fillna(media_precio)

print(df.isnull().sum())
```

```python
df.head()
```

```python
df_competencia = pd.read_csv('../data/scrap_competencia.csv')
df_competencia
```

```python
df.to_csv('../data/reservas_hoteles.csv')
```

```python
id_hotel = df[df['competencia'] == True]['id_hotel'].unique()
```

```python
id_hotel
```

```python
df_competencia['id_hotel'] = id_hotel
```

```python
df_competencia
```

```python
df_final = pd.merge(df,df_competencia, on='id_hotel', how='left', suffixes=('', '_nuevo'))
```

```python
df_final
```

```python
cols_a_actualizar = ['precio_noche', 'nombre_hotel', 'estrellas']
for col in cols_a_actualizar:
    df.loc[df['competencia'], col] = df_final.loc[df['competencia'], col + '_nuevo']
```

```python
df.head()
```

```python
df.info()
```

```python
df['fecha_reserva'] = df['fecha_reserva'].replace('', '2025-02-19')
df.head()
```

```python
ids_repetidos = df.groupby('nombre_hotel')['id_hotel'].unique().reset_index()
ids_repetidos
```

```python
df['id_hotel'] = pd.factorize(df['nombre_hotel'])[0] + 1
df
```

```python
df['numero'] = df.groupby('mail').ngroup() + 1
df['id_cliente'] = df['competencia'].map({True: 'HC', False: 'HP'}) + df['numero'].astype(str)
df.drop(columns=['numero'], inplace=True)
df.info()
```

```python
df.to_csv('../data/reservas_hoteles.csv')
df.to_parquet('../data/reservas_hoteles_limpio.parquet')
```
